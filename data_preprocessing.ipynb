```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('./personality_datasert.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.loc[df['Drained_after_socializing']=="Yes",'Drained_after_socializing']=1
df.loc[df['Drained_after_socializing']=="No",'Drained_after_socializing']=0
```

```python
df.loc[df['Stage_fear']=="Yes",'Stage_fear']=1
df.loc[df['Stage_fear']=="No",'Stage_fear']=0
```

```python
df.loc[df['Personality']=="Extrovert",'Personality']=0
df.loc[df['Personality']=="Introvert",'Personality']=1
```

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
X_train
```

```python
y_train
```

```python
merged_train = pd.concat([X_train, y_train], axis=1)
merged_train
```

```python
merged_test = pd.concat([X_test, y_test], axis=1)
merged_test
```

```python
merged_train.to_csv('./train.csv')
merged_test.to_csv('./test.csv')
```
